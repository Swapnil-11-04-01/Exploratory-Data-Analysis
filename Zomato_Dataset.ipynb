```python
df_country = pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df, df_country, on = "Country Code", how = "left")
```

```python
final_df.head(2)
```

```python
##To Check Datatypes
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
final_df.Country.value_counts().index
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
##Pie CHart
plt.pie(country_val, labels = country_names)
```

```python
##Pie CHart - Top 3 Countries That Uses Zomato
plt.pie(country_val[:3], labels = country_names[:3])
```

```python
plt.pie(country_val[:3], labels = country_names[:3], autopct = "%1.2f%%")
```

Observation : Zomato's maximum records/transactions are from India, then United Staes and then United Kingdom

```python
final_df.columns
```

```python
final_df.groupby(["Aggregate rating", "Rating color", "Rating text"]).size()
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

**Observation**

1.   When Ratings are betwee 4.5 to 4.9 -----> Excellent
2.   When Ratings are between 4.0 to 4.4 -----> Very Good
3.   When Ratings are between 3.5 to 3.9 -----> Good
4.   When Ratings are between 2.5 to 3.4 -----> Average
5.   When Ratings are between 2.0 to 2.4 -----> Poor


```python
ratings.head()
```

```python
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count',hue = "Rating color", data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count',hue = "Rating color", data=ratings, palette=['white', 'red', 'orange', 'yellow', 'green', 'green'])
```

**Observations**

1.   Not Rated count is very high
2.   Maximum number of ratings are between 2.5 to 3.4



```python
##Count Plot
sns.countplot(x='Rating color',data=ratings, palette=['white', 'red', 'orange', 'yellow', 'green', 'green'])
```

```python
### Find the countries name that has given 0 ratings
final_df[final_df["Rating color"]=="White"].groupby("Country").size().reset_index()
```

```python
##Find out which currency is used by which Country.
final_df[['Country', 'Currency']].groupby(['Country', 'Currency']).size().reset_index()
```

```python
##Which countries do have online deliveries option?
final_df[final_df["Has Online delivery"] == "Yes"].groupby("Country").size().reset_index()
```

**Observations:**

1.   Online deliveries are available in India & UAE
